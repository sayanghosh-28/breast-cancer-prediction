```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("/content/breast_cancer.csv")
data.head()
```

```python
data['diagnosis'].value_counts()
```

```python
data.info()
```

```python
data=data.drop(['id','Unnamed: 32'],axis=1)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['diagnosis']=encoder.fit_transform(data['diagnosis'])
data.head()
```

```python
x=data.drop(['diagnosis'],axis=1)
y=data['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
### Split into train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
## Neural Network
```

```python
import tensorflow.keras as tf
```

```python
model=tf.models.Sequential()
model.add(tf.layers.Dense(16,input_dim=30,activation='relu'))
model.add(tf.layers.Dense(32,activation='relu'))      ## hidden
model.add(tf.layers.Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=["accuracy"])
```

```python
model.fit(xtrain,ytrain,epochs=500)
```

```python
ytest
```

```python
ypred=model.predict(xtest)
```

```python
ypred=ypred.round()
ypred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(ytest,ypred)
cm
```

```python
accuracy=accuracy_score(ytest,ypred)
accuracy
```
